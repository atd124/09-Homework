```python
# Another option is the POST request
# the info is..... not in the URL.
# you can't use requests.get!
# you need to use requests.post! + give extra data
import requests
from bs4 import BeautifulSoup

response = requests.post("https://fdic-search.app.cloud.gov/")
doc = BeautifulSoup(response.text)
page_content = doc.select_one(".main-results-without-dn")

print(page_content.text)
```
