```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# Send headers to fake being a browser a little bit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}

response = requests.get("https://chartable.com/charts/chartable", headers=headers)
doc = BeautifulSoup(response.text)
doc
```

```python
podcasts = []

table = doc.find("table")

for row in table.select('tr'):
    showtitle = (row.find("a", class_="link blue").text)
    showurl = (row.find("a", class_="link blue")['href'])
    rank = (row.find("div", class_="b header-font f2 tc").text)
    network = (row.select_one(".mb1.db").text)
```

```python
podcasts = []

table = doc.find("table")

for row in table.select('tr'):
    showtitle = (row.find("a", class_="link blue").text)
    showurl = (row.find("a", class_="link blue")['href'])
    rank = (row.find("div", class_="b header-font f2 tc").text)
    network = (row.select_one(".mb1.db").text)
    
    podcast_dict = {
        'rank': rank,
        'title': showtitle,
        'url': showurl,
        'network': network
    }
    
    podcasts.append(podcast_dict)
    
df= pd.DataFrame(podcasts)
df.to_csv('podcasts.csv', index=False)
df
```
